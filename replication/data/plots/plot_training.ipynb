```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import os
import argparse

import numpy as np
```

```python
# Plot training and test loss, and training and test accuracy
def plot(title, fn, save_name=None):
    metrics_train = np.load(fn + "metrics_train.npy")
    metrics_test = np.load(fn + "metrics_test.npy")
    # Final test accuracy
    print("Final test accuracy: {:.2f}%".format(metrics_test[-1,1]))
    fig, ax = plt.subplots(1,2, figsize=(10,5))
    epochs = len(metrics_train)
    ax[0].plot(range(epochs), metrics_train[:,0], label="train")
    ax[0].plot(range(epochs), metrics_test[:,0], label="test")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Loss")
    ax[0].legend()

    ax[1].plot(range(epochs), metrics_train[:,1], label="train")
    ax[1].plot(range(epochs), metrics_test[:,1], label="test")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Accuracy (%)")
    ax[1].set_ylim([0,100])
    ax[1].legend()

    fig.suptitle(title)
    
    if save_name is not None:
        plt.savefig(save_name)
```

```python
fn = "../cifar10/resnet20_s77_ep200_lr0.001_bs128_wd0.0001_mo0.9_losscross_entropy/"

plot("ResNet20 on CIFAR10 (CE Loss)", fn, save_name="resnet20_training_ce.png")
```

```python
fn = "../cifar10/resnet20_s0_ep200_lr0.001_bs128_wd0.0001_mo0.9_lossreward/"

plot("ResNet20 on CIFAR10 (Reward Loss)", fn, save_name="resnet20_training_reward.png")
```
